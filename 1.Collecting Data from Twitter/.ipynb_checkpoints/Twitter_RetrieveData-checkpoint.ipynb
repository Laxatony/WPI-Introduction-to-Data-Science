```python
# Transform data into JSON format
import json
text = json.dumps([status._json for status in wpi])
jsonData = json.loads(text)
jsonData
```

```python
# Print the username and the content for each tweeet
for tweet in jsonData:
    print('%s: %s\n' % (tweet['user']['name'], tweet['text']))
```
